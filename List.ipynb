```python
List = [1, 5, 23, 2, 8]
print(List)
```

```python
List.append(9)
print('Appended List',List)
```

```python
# .insert(index, element)
List.insert(3, 12)
print('Insertion in List',List)
```

```python
List.remove(23)
print('After Removal',List)
```

```python
# .pop(index) - if index not defined remove last item
last = List.pop()
print('Last Item',last)
```

```python
print('Count of 2 in List: ', List.count(2))
```

```python
copiedList = List.copy()
print('Copied List: ', copiedList)
```

```python
# .reverse()
List.reverse()
print('Reverse List: ', List)
```

```python
# .sort()
List.sort()
print('Sorted List: ', List)
```

```python
List.sort(reverse=True)
print('Sorted List in Descending Order: ', List)
```

```python
sumList = sum(List)
print('Sum of List: ', sumList)
```
