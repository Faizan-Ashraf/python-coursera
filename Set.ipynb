```python
A = {1, 3 ,6 ,8 ,9}
B = {3, 8, 1, 54, 2, 12}
```

```python
print(A.union(B))
# same as
print(A | B)
```

```python
print(A.intersection(B))
# same as
print(A & B)
```

```python
print(A.difference(B))
# same as
print(A - B)
```

```python
print(A.issubset(B))
print(A.issuperset(B))
print(A.isdisjoint(B))
```
