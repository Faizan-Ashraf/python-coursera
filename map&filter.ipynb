```python
coffee = ["expresso", "mosha", "latte", "cappuccino", "cortado" , "americano"]

def coffee_with_c(coffee):
  if coffee[0] == 'c':
    return coffee
```

### **1. Map Function**

```python
# Map Takes all objects in a list and applies a function
c_coffees = map(coffee_with_c, coffee)
print(list(c_coffees))
```

### **2. Filter Function**

```python
# filters do the same, but take the results and creates a new list  with  only the true values
c_coffee = filter(coffee_with_c, coffee)
print(list(c_coffee))
```
