```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("tips.csv")
```

----
# **Total Bill Distribution:**

Question: Create a histogram to visualize the distribution of total_bill.<br>
Task: Customize the number of bins and add labels to the axes and a title. Discuss the shape of the distribution.

```python
sns.histplot(x='total_bill', data=df, bins=20, kde=True)
```

# **Tip vs. Total Bill:**

Question: Create a scatter plot to visualize the relationship between total_bill and tip.<br>
Task: Use different colors to represent the sex of the payer. Discuss any patterns observed.

```python
sns.scatterplot(x='total_bill',y='tip',data=df,hue='sex')
```

# **Average Tip by Gender:**

Question: Create a bar plot to show the average tip amount given by sex.<br>
Task: Annotate the bars with the exact average tip values.

```python
sns.barplot(x='sex',y='tip',data=df)
```

# **Total Bill by Day:**

Question: Generate a box plot to visualize the distribution of total_bill for each day of the week.<br>
Task: Discuss how the total bill varies by day and any potential outliers.

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

# **Size of Party vs. Total Bill:**

Question: Create a scatter plot to examine the relationship between size (size of the party) and total_bill.<br>
Task: Use different marker shapes or colors to indicate whether the party was a smoker or non-smoker.

```python
sns.scatterplot(x='size',y='total_bill',data=df, style='smoker',hue='smoker')
```

# **Price Per Person Analysis:**

Question: Create a violin plot to visualize the distribution of price_per_person across different days.<br>
Task: Discuss the differences in spending per person by day.

```python
sns.violinplot(x='day',y='price_per_person',data=df, palette="Set2")
```

# **Tips by Time of Day:**

Question: Create a box plot to compare the distribution of tip amounts for Dinner and Lunch.<br>
Task: Discuss any differences in tipping behavior based on the time of day.

```python
sns.boxplot(x='time',y='tip',data=df)
```

# **Correlation Analysis:**

Question: Generate a heatmap to visualize the correlation between numerical variables in the dataset (e.g., total_bill, tip, size, price_per_person).<br>
Task: Discuss any strong correlations found between the variables.

```python
data = df[['total_bill', 'tip', 'size', 'price_per_person']]
correlation = data.corr()

sns.heatmap(correlation, cmap='coolwarm')
```

## **Smoker vs. Non-Smoker Spending:**

Question: Create a bar plot showing the average total_bill for smokers vs. non-smokers.<br>
Task: Discuss how smoking status might influence spending.

```python
sns.barplot(x='smoker',y='total_bill',data=df)
```

# **Pie Chart of Payment Methods:**

Question: Create a pie chart showing the proportion of sex among the payers.<br>
Task: Discuss the implications of the observed proportions in the context of the dataset.

```python
sns.countplot(x='sex',data=df, palette="Set2")
```
