```python
student = {
    'name': 'faizan',
    'age': 22,
    'courses': ['Math', 'CS']
}
print(student)
print(student['name'])
print(student['age'])
print(student['courses'])
```

```python
# keys() return all keys
print(student.keys())
```

```python
# values() return all values
print(student.values())
```

```python
# .items() â€“ Returns a view of all key-value pairs as tuples.
print(student.items())
```

```python
student.update({"age": 23, "city": "New York"})
print(student)
```

```python
student.setdefault("age", 25)
print(student)  # If 'age' wasn't present, it would be set to 25
```

```python
student.clear()
print(student)  # Output: {}
```
