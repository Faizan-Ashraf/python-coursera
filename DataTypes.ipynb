### **1. Numeric**

```python
integer = 24
floatdt = 2.4
cnum= 10 + 10j

print(integer)
print(floatdt)
print(cnum)

print(type(integer))
print(type(floatdt))
print(type(cnum))
```

### **2. Sequence**

```python
string = 'faizan'
List = [1, 'faizan', True] # They are essentially an array and hold any type inside square brackets.
typle = (1, 'faizan', True) #tuples are similar to list but they are immutable - values cannot be changed

print(string)
print(List)
print(typle)

print(type(string))
print(type(List))
print(type(typle))
```

### **3. Dictionary**

```python
# Dictionary stores data in key-value pairs with any DataType
data = { 'Name': 'Faizan', 'Age': '22'}
print(data)
print(type(data))
```

### **3. Boolean**

```python
# boolean is either true or false
flag = True
print(flag)
print(type(flag))
```

### **4. Set**

```python
# set is an unordered and non-indexed collection of non-repeated values
Set = {1, 2, 'faizan', 'True', 5, 5}
print(Set)
print(type(Set))
```
