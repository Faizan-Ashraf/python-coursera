### **Args**

```python
# Allows a function to accept any number of positional arguments as tuple.
def sumofvalues(*args):
  return sum(args)

print(sumofvalues(2, 6, 8))
```

### **Kwargs**

```python
# Allows a function to accept any number of keyword arguments (key=value pairs) as dictionary.
def total(**kwargs):
  sum = 0
  for key, value in kwargs.items():
    sum += value
  return sum

print(total(coffee = 2, juice = 4))
```
