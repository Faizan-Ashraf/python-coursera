### **__init__ Function**
This function executes when the class initiated like a defaault constructor.

```python
class person:
  def __init__(self, name, age) : # The 'self' parameter is a reference to the current instance of the class
    self.name = name
    self.age = age

p1 = person("John", 36)

print(p1.name)
print(p1.age)
```

### **__str__() Function**
The __str__() function controls what should be returned when the class object is represented as a string.

```python
class person:
  def __init__(self, name, age) :
    self.name = name
    self.age = age

  def __str__(self) :
    return f"{self.name}({self.age})"

p1 = person("John", 36)
print(p1)

del p1
```
