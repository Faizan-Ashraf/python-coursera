### **Ecapsulation**

```python
class person:
  def __init__(self, name, age) :
    self.name = name #public
    self.__age = age #private

  def __str__(self) :
    return f"{self.name}({self.__age})"

  def getAge(self): #getter for private attributes
      return self.__age

  def setAge(self, age): #setter for private attributes
      self.__age = age

p1 = person("John", 36)
print(p1.name)
print(p1.getAge())
print(p1)

del p1
```
