```python
List = [2, 6, 8]

try:
  print(List[3])
except IndexError as e:
  print(e)
```

```python
def divide_by(a, b):
    return a / b

try:
    ans = divide_by(40, 0)
    print(ans)
except ZeroDivisionError as e:
    print(e)
```
