### **Single Inheritance**

```python
class Animal:
    def speak(self):
        return "Animal makes a sound"

class Dog(Animal):  # Dog inherits from Animal
    def speak(self):
        return "Bark"

dog = Dog()
print(dog.speak())
```

### **Multiple Inheritance**

```python
class Animal:
    def eat(self):
        return "Eating..."

class Mammal:
    def warm_blooded(self):
        return True

class Dog(Animal, Mammal):  # Inheriting from both Animal & Mammal
    def speak(self):
        return "Bark"


dog = Dog()
print(dog.eat())
print(dog.warm_blooded())
print(dog.speak())
```

### **Multilevel Inheritance**

```python
class Animal:
    def breathe(self):
        return "Breathing..."

class Mammal(Animal):
    def walk(self):
        return "Walking..."

class Dog(Mammal):
    def speak(self):
        return "Bark"

# Creating an object
dog = Dog()
print(dog.breathe())
print(dog.walk())
print(dog.speak())
```

### **Hierarchical Inheritance**

```python
class Animal:
    def speak(self): #polymorphism
        return "Some sound"

class Dog(Animal):
    def speak(self):
        return "Bark"

class Cat(Animal):
    def speak(self):
        return "Meow"

# Creating objects
dog = Dog()
cat = Cat()
print(dog.speak())
print(cat.speak())
```
